# Single-Qubit Clifford Gates: NMR Pulse Decompositions

The 24 single-qubit Clifford gates correspond to the 24 rotational symmetries of the octahedron (equivalently, the cube). They are the unitaries $U$ (up to global phase) that map Pauli operators to Pauli operators under conjugation: $U \sigma_\alpha U^\dagger \in \{\pm\sigma_x, \pm\sigma_y, \pm\sigma_z\}$.

We decompose each into NMR-realizable pulse sequences using:
- $R_x(\theta) = e^{-i\theta\sigma_x/2}$ — RF pulse along $x$
- $R_y(\theta) = e^{-i\theta\sigma_y/2}$ — RF pulse along $y$
- $R_z(\theta) = e^{-i\theta\sigma_z/2}$ — Free evolution / frame shift

## The 24 Single-Qubit Cliffords

Organized by geometric class (rotational symmetries of the octahedron):

| # | Name | $X \to$ | $Y \to$ | $Z \to$ | Pulse Sequence | Class |
|---|------|---------|---------|---------|----------------|-------|
| 0 | $I$ | $+X$ | $+Y$ | $+Z$ | $I$ | Identity |
| 1 | $X$ | $+X$ | $-Y$ | $-Z$ | $R_x(\pi)$ | $\pi$ about axis |
| 2 | $Y$ | $-X$ | $+Y$ | $-Z$ | $R_y(\pi)$ | $\pi$ about axis |
| 3 | $Z$ | $-X$ | $-Y$ | $+Z$ | $R_z(\pi)$ | $\pi$ about axis |
| 4 | $\sqrt{X}$ | $+X$ | $+Z$ | $-Y$ | $R_x(\pi/2)$ | $\pm\pi/2$ about axis |
| 5 | $\sqrt{X}^\dagger$ | $+X$ | $-Z$ | $+Y$ | $R_x(-\pi/2)$ | $\pm\pi/2$ about axis |
| 6 | $\sqrt{Y}$ | $-Z$ | $+Y$ | $+X$ | $R_y(\pi/2)$ | $\pm\pi/2$ about axis |
| 7 | $\sqrt{Y}^\dagger$ | $+Z$ | $+Y$ | $-X$ | $R_y(-\pi/2)$ | $\pm\pi/2$ about axis |
| 8 | $S = \sqrt{Z}$ | $+Y$ | $-X$ | $+Z$ | $R_z(\pi/2)$ | $\pm\pi/2$ about axis |
| 9 | $S^\dagger$ | $-Y$ | $+X$ | $+Z$ | $R_z(-\pi/2)$ | $\pm\pi/2$ about axis |
| 10 | $H$ | $+Z$ | $-Y$ | $+X$ | $R_y(\pi/2) \cdot R_z(\pi)$ | $\pi$ about face diagonal |
| 11 | $H_{\overline{xz}}$ | $-Z$ | $-Y$ | $-X$ | $R_y(-\pi/2) \cdot R_z(\pi)$ | $\pi$ about face diagonal |
| 12 | $H_{yz}$ | $-X$ | $+Z$ | $+Y$ | $R_x(-\pi/2) \cdot R_z(\pi)$ | $\pi$ about face diagonal |
| 13 | $H_{\overline{yz}}$ | $-X$ | $-Z$ | $-Y$ | $R_x(\pi/2) \cdot R_z(\pi)$ | $\pi$ about face diagonal |
| 14 | $H_{xy}$ | $+Y$ | $+X$ | $-Z$ | $R_z(\pi/2) \cdot R_x(\pi)$ | $\pi$ about face diagonal |
| 15 | $H_{\overline{xy}}$ | $-Y$ | $-X$ | $-Z$ | $R_z(-\pi/2) \cdot R_x(\pi)$ | $\pi$ about face diagonal |
| 16 | $V_1$ | $+Y$ | $+Z$ | $+X$ | $R_y(\pi/2) \cdot R_z(\pi/2)$ | $2\pi/3$ about body diagonal |
| 17 | $V_1^\dagger$ | $+Z$ | $+X$ | $+Y$ | $R_z(-\pi/2) \cdot R_y(-\pi/2)$ | $-2\pi/3$ about body diagonal |
| 18 | $V_2$ | $-Z$ | $+X$ | $-Y$ | $R_z(-\pi/2) \cdot R_y(\pi/2)$ | $2\pi/3$ about body diagonal |
| 19 | $V_2^\dagger$ | $+Y$ | $-Z$ | $-X$ | $R_y(-\pi/2) \cdot R_z(\pi/2)$ | $-2\pi/3$ about body diagonal |
| 20 | $V_3$ | $-Z$ | $-X$ | $+Y$ | $R_x(-\pi/2) \cdot R_z(\pi/2)$ | $2\pi/3$ about body diagonal |
| 21 | $V_3^\dagger$ | $-Y$ | $+Z$ | $-X$ | $R_z(-\pi/2) \cdot R_x(\pi/2)$ | $-2\pi/3$ about body diagonal |
| 22 | $V_4$ | $+Z$ | $-X$ | $-Y$ | $R_y(-\pi/2) \cdot R_x(\pi/2)$ | $2\pi/3$ about body diagonal |
| 23 | $V_4^\dagger$ | $-Y$ | $-Z$ | $+X$ | $R_x(-\pi/2) \cdot R_y(\pi/2)$ | $-2\pi/3$ about body diagonal |

**Geometric classes:**
- **Identity** (1): trivial
- **$\pi$ about coordinate axes** (3): $X$, $Y$, $Z$ gates
- **$\pm\pi/2$ about coordinate axes** (6): $\sqrt{X}$, $\sqrt{Y}$, $S$ and their adjoints
- **$\pi$ about face-diagonal axes** (6): Hadamard-class gates — $\pi$ rotations about axes like $(\hat{x}+\hat{z})/\sqrt{2}$
- **$\pm 2\pi/3$ about body-diagonal axes** (8): $2\pi/3$ rotations about axes like $(\hat{x}+\hat{y}+\hat{z})/\sqrt{3}$

Total: $1 + 3 + 6 + 6 + 8 = 24$ ✓

```python
import numpy as np
from qutip import Qobj, qeye, sigmax, sigmay, sigmaz

pi = np.pi
sx, sy, sz, si = sigmax(), sigmay(), sigmaz(), qeye(2)

PAULIS = {'I': si, 'X': sx, 'Y': sy, 'Z': sz,
          '-X': -sx, '-Y': -sy, '-Z': -sz}
```

```python
def Rx(theta):
    return (-1j * theta / 2 * sx).expm()

def Ry(theta):
    return (-1j * theta / 2 * sy).expm()

def Rz(theta):
    return (-1j * theta / 2 * sz).expm()

def R_axis(n, theta):
    """Rotation by theta about unit axis n = (nx, ny, nz)."""
    nx, ny, nz = n
    H = nx * sx + ny * sy + nz * sz
    return (-1j * theta / 2 * H).expm()

def phase_equiv(U, V, tol=1e-9):
    """Check if U and V are equal up to a global phase."""
    overlap = (U.dag() * V).tr()
    return abs(abs(overlap) - 2) < tol

def clean(M):
    arr = np.vectorize(lambda x: complex(round(x.real, 4), round(x.imag, 4)))(M.full())
    return Qobj(arr)
```

## Define All 24 Cliffords

Each Clifford is specified by:
- `name`: human-readable label
- `pulse_seq`: the NMR pulse sequence (product of rotations, applied right-to-left)
- `target`: the expected unitary from the axis/angle formula: $R_{\hat{n}}(\theta) = e^{-i\theta(\hat{n}\cdot\vec{\sigma})/2}$
- `pauli_map`: expected Heisenberg-picture mapping $(X\to, Y\to, Z\to)$

```python
s3 = 1 / np.sqrt(3)
s2 = 1 / np.sqrt(2)

cliffords = [
    # --- Identity (1) ---
    dict(name='I',        pulse=lambda: si,                     axis=None,           angle=0,       x='X',  y='Y',  z='Z'),

    # --- pi about coordinate axes (3) ---
    dict(name='X',        pulse=lambda: Rx(pi),                 axis=(1,0,0),        angle=pi,      x='X',  y='-Y', z='-Z'),
    dict(name='Y',        pulse=lambda: Ry(pi),                 axis=(0,1,0),        angle=pi,      x='-X', y='Y',  z='-Z'),
    dict(name='Z',        pulse=lambda: Rz(pi),                 axis=(0,0,1),        angle=pi,      x='-X', y='-Y', z='Z'),

    # --- ±pi/2 about coordinate axes (6) ---
    dict(name='√X',       pulse=lambda: Rx(pi/2),               axis=(1,0,0),        angle=pi/2,    x='X',  y='Z',  z='-Y'),
    dict(name='√X†',      pulse=lambda: Rx(-pi/2),              axis=(1,0,0),        angle=-pi/2,   x='X',  y='-Z', z='Y'),
    dict(name='√Y',       pulse=lambda: Ry(pi/2),               axis=(0,1,0),        angle=pi/2,    x='-Z', y='Y',  z='X'),
    dict(name='√Y†',      pulse=lambda: Ry(-pi/2),              axis=(0,1,0),        angle=-pi/2,   x='Z',  y='Y',  z='-X'),
    dict(name='S',        pulse=lambda: Rz(pi/2),               axis=(0,0,1),        angle=pi/2,    x='Y',  y='-X', z='Z'),
    dict(name='S†',       pulse=lambda: Rz(-pi/2),              axis=(0,0,1),        angle=-pi/2,   x='-Y', y='X',  z='Z'),

    # --- pi about face-diagonal axes (6) ---
    dict(name='H',        pulse=lambda: Ry(pi/2) * Rz(pi),     axis=(s2,0,s2),      angle=pi,      x='Z',  y='-Y', z='X'),
    dict(name='H_xz̄',    pulse=lambda: Ry(-pi/2) * Rz(pi),    axis=(s2,0,-s2),     angle=pi,      x='-Z', y='-Y', z='-X'),
    dict(name='H_yz',     pulse=lambda: Rx(-pi/2) * Rz(pi),    axis=(0,s2,s2),      angle=pi,      x='-X', y='Z',  z='Y'),
    dict(name='H_yz̄',    pulse=lambda: Rx(pi/2) * Rz(pi),     axis=(0,s2,-s2),     angle=pi,      x='-X', y='-Z', z='-Y'),
    dict(name='H_xy',     pulse=lambda: Rz(pi/2) * Rx(pi),     axis=(s2,s2,0),      angle=pi,      x='Y',  y='X',  z='-Z'),
    dict(name='H_xȳ',    pulse=lambda: Rz(-pi/2) * Rx(pi),    axis=(s2,-s2,0),     angle=pi,      x='-Y', y='-X', z='-Z'),

    # --- ±2pi/3 about body-diagonal axes (8) ---
    dict(name='V₁',       pulse=lambda: Ry(pi/2) * Rz(pi/2),   axis=(s3,s3,s3),     angle=2*pi/3,  x='Y',  y='Z',  z='X'),
    dict(name='V₁†',      pulse=lambda: Rz(-pi/2) * Ry(-pi/2), axis=(s3,s3,s3),     angle=-2*pi/3, x='Z',  y='X',  z='Y'),
    dict(name='V₂',       pulse=lambda: Rz(-pi/2) * Ry(pi/2),  axis=(s3,s3,-s3),    angle=2*pi/3,  x='-Z', y='X',  z='-Y'),
    dict(name='V₂†',      pulse=lambda: Ry(-pi/2) * Rz(pi/2),  axis=(s3,s3,-s3),    angle=-2*pi/3, x='Y',  y='-Z', z='-X'),
    dict(name='V₃',       pulse=lambda: Rx(-pi/2) * Rz(pi/2),  axis=(-s3,s3,s3),    angle=2*pi/3,  x='-Z', y='-X', z='Y'),
    dict(name='V₃†',      pulse=lambda: Rz(-pi/2) * Rx(pi/2),  axis=(-s3,s3,s3),    angle=-2*pi/3, x='-Y', y='Z',  z='-X'),
    dict(name='V₄',       pulse=lambda: Ry(-pi/2) * Rx(pi/2),  axis=(s3,-s3,s3),    angle=2*pi/3,  x='Z',  y='-X', z='-Y'),
    dict(name='V₄†',      pulse=lambda: Rx(-pi/2) * Ry(pi/2),  axis=(s3,-s3,s3),    angle=-2*pi/3, x='-Y', y='-Z', z='X'),
]

assert len(cliffords) == 24, f'Expected 24, got {len(cliffords)}'
```

## Validation 1: Pulse Sequence vs Axis-Angle Target

For each Clifford, verify that the pulse sequence produces the same unitary as $R_{\hat{n}}(\theta) = e^{-i\theta(\hat{n}\cdot\vec{\sigma})/2}$ (up to global phase).

```python
print('Validation 1: Pulse sequence == axis/angle target (up to global phase)')
print('=' * 72)

all_pass_1 = True
for i, c in enumerate(cliffords):
    U_pulse = c['pulse']()
    if c['axis'] is not None:
        U_target = R_axis(c['axis'], c['angle'])
    else:
        U_target = si

    ok = phase_equiv(U_pulse, U_target)
    status = '✓' if ok else '✗ FAIL'
    print(f"  C{i:2d}  {c['name']:6s}  {status}")
    if not ok:
        all_pass_1 = False
        print(f'       Pulse: {clean(U_pulse)}')
        print(f'       Target: {clean(U_target)}')

print('\n' + ('ALL PASSED ✓' if all_pass_1 else 'SOME FAILED ✗'))
```

## Validation 2: Heisenberg-Picture Pauli Mapping

For each Clifford $C$, verify that $C \sigma_\alpha C^\dagger$ matches the expected Pauli mapping.

```python
def identify_pauli(M, tol=1e-9):
    """Identify which ±Pauli a matrix is (or return None)."""
    for label, P in PAULIS.items():
        if (M - P).norm() < tol:
            return label
    return None

print('Validation 2: Pauli conjugation mapping')
print('=' * 72)

all_pass_2 = True
for i, c in enumerate(cliffords):
    U = c['pulse']()
    results = {}
    ok = True
    for pauli_name, pauli_op, expected_key in [('X', sx, 'x'), ('Y', sy, 'y'), ('Z', sz, 'z')]:
        conjugated = U * pauli_op * U.dag()
        label = identify_pauli(conjugated)
        results[pauli_name] = label
        if label != c[expected_key]:
            ok = False

    status = '✓' if ok else '✗ FAIL'
    print(f"  C{i:2d}  {c['name']:6s}  X→{results['X']:3s}  Y→{results['Y']:3s}  Z→{results['Z']:3s}  {status}")
    if not ok:
        all_pass_2 = False
        print(f"       Expected: X→{c['x']}  Y→{c['y']}  Z→{c['z']}")

print('\n' + ('ALL PASSED ✓' if all_pass_2 else 'SOME FAILED ✗'))
```

## Validation 3: Uniqueness — All 24 Are Distinct

Verify no two Cliffords are the same (up to global phase).

```python
print('Validation 3: Uniqueness check')
print('=' * 72)

unitaries = [c['pulse']() for c in cliffords]
duplicates = []
for i in range(24):
    for j in range(i + 1, 24):
        if phase_equiv(unitaries[i], unitaries[j]):
            duplicates.append((i, j))

if duplicates:
    print('DUPLICATES FOUND ✗:')
    for i, j in duplicates:
        print(f'  C{i} ({cliffords[i]["name"]}) == C{j} ({cliffords[j]["name"]})')
else:
    print(f'All 24 Cliffords are distinct ✓')
```

## Validation 4: Completeness — Generators Produce All 24

The single-qubit Clifford group is generated by $H$ and $S$. Verify that composing $H$ and $S$ can produce all 24 elements.

```python
print('Validation 4: Completeness via generators H, S')
print('=' * 72)

H_gate = cliffords[10]['pulse']()
S_gate = cliffords[8]['pulse']()

generated = {0: si}  # start with identity
queue = [si]
max_depth = 10

# BFS over products of H and S
for _ in range(max_depth):
    new_queue = []
    for U in queue:
        for gen in [H_gate, S_gate]:
            V = gen * U
            is_new = True
            for existing in generated.values():
                if phase_equiv(V, existing):
                    is_new = False
                    break
            if is_new:
                idx = len(generated)
                generated[idx] = V
                new_queue.append(V)
    queue = new_queue
    if len(generated) == 24:
        break

print(f'Generated {len(generated)} unique Cliffords from H and S')

# check every generated element matches one of our 24
matched = set()
for V in generated.values():
    for i, c in enumerate(cliffords):
        if phase_equiv(V, c['pulse']()):
            matched.add(i)
            break

if matched == set(range(24)):
    print('All 24 Cliffords are reachable from {H, S} ✓')
else:
    missing = set(range(24)) - matched
    print(f'Missing Cliffords: {[cliffords[i]["name"] for i in missing]} ✗')
```

## Validation 5: Group Closure

Verify that the product of any two Cliffords is again one of the 24.

```python
print('Validation 5: Group closure (24 × 24 = 576 products)')
print('=' * 72)

unitaries = [c['pulse']() for c in cliffords]
closure_ok = True
failures = 0

for i in range(24):
    for j in range(24):
        product = unitaries[i] * unitaries[j]
        found = False
        for k in range(24):
            if phase_equiv(product, unitaries[k]):
                found = True
                break
        if not found:
            closure_ok = False
            failures += 1
            if failures <= 5:
                print(f'  C{i} * C{j} not in group! ✗')

if closure_ok:
    print('All 576 products land in the group ✓')
else:
    print(f'{failures} products failed closure ✗')
```

## Validation 6: Inverses Exist

For each Clifford, verify it has an inverse in the group.

```python
print('Validation 6: Inverses')
print('=' * 72)

unitaries = [c['pulse']() for c in cliffords]
all_pass_6 = True

for i in range(24):
    U_inv = unitaries[i].dag()
    found_inv = None
    for j in range(24):
        if phase_equiv(U_inv, unitaries[j]):
            found_inv = j
            break
    if found_inv is not None:
        print(f"  C{i:2d} ({cliffords[i]['name']:6s})⁻¹ = C{found_inv:2d} ({cliffords[found_inv]['name']})")
    else:
        print(f"  C{i:2d} ({cliffords[i]['name']:6s}) has no inverse in the set! ✗")
        all_pass_6 = False

print('\n' + ('ALL PASSED ✓' if all_pass_6 else 'SOME FAILED ✗'))
```

## Display: Unitary Matrices

Show the matrix representation of each Clifford (from the pulse sequence).

```python
for i, c in enumerate(cliffords):
    U = c['pulse']()
    print(f"C{i:2d}: {c['name']}")
    display(clean(U))
```
