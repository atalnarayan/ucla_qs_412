# Single-Qubit Clifford Gates: NMR Pulse Decompositions

The 24 single-qubit Clifford gates correspond to the 24 rotational symmetries of the octahedron (equivalently, the cube). They are the unitaries $U$ (up to global phase) that map Pauli operators to Pauli operators under conjugation: $U \sigma_\alpha U^\dagger \in \{\pm\sigma_x, \pm\sigma_y, \pm\sigma_z\}$.

We provide **two decompositions** for each gate:
1. **With $R_z$**: Using $R_x$, $R_y$, $R_z$ вЂ” suitable when frame shifts or free evolution are available
2. **$R_x/R_y$ only**: Using only RF pulses вЂ” the hardware-native NMR gate set


Primitive gates:The $R_z$-elimination identity is: $R_z(\theta) = R_x(\pi/2) \cdot R_y(\theta) \cdot R_x(-\pi/2)$

- $R_x(\theta) = e^{-i\theta\sigma_x/2}$ вЂ” RF pulse along $x$

- $R_y(\theta) = e^{-i\theta\sigma_y/2}$ вЂ” RF pulse along $y$- $R_z(\theta) = e^{-i\theta\sigma_z/2}$ вЂ” Free evolution / frame shift

**Pulse count distribution ($R_x/R_y$ only):** 0 pulses (1), 1 pulse (6), 2 pulses (12), 3 pulses (5)

**Geometric classes:**
- **Identity** (1): trivial
- **$\pi$ about coordinate axes** (3): $X$, $Y$, $Z$ gates
- **$\pm\pi/2$ about coordinate axes** (6): $\sqrt{X}$, $\sqrt{Y}$, $S$ and their adjoints
- **$\pi$ about face-diagonal axes** (6): Hadamard-class gates вЂ” $\pi$ rotations about axes like $(\hat{x}+\hat{z})/\sqrt{2}$
- **$\pm 2\pi/3$ about body-diagonal axes** (8): $2\pi/3$ rotations about axes like $(\hat{x}+\hat{y}+\hat{z})/\sqrt{3}$

Total: $1 + 3 + 6 + 6 + 8 = 24$ вњ“

```python
import numpy as np
from qutip import Qobj, qeye, sigmax, sigmay, sigmaz

pi = np.pi
sx, sy, sz, si = sigmax(), sigmay(), sigmaz(), qeye(2)

PAULIS = {'I': si, 'X': sx, 'Y': sy, 'Z': sz,
          '-X': -sx, '-Y': -sy, '-Z': -sz}
```

```python
def Rx(theta):
    return (-1j * theta / 2 * sx).expm()

def Ry(theta):
    return (-1j * theta / 2 * sy).expm()

def Rz(theta):
    return (-1j * theta / 2 * sz).expm()

def R_axis(n, theta):
    """Rotation by theta about unit axis n = (nx, ny, nz)."""
    nx, ny, nz = n
    H = nx * sx + ny * sy + nz * sz
    return (-1j * theta / 2 * H).expm()

def phase_equiv(U, V, tol=1e-9):
    """Check if U and V are equal up to a global phase."""
    overlap = (U.dag() * V).tr()
    return abs(abs(overlap) - 2) < tol

def clean(M):
    arr = np.vectorize(lambda x: complex(round(x.real, 4), round(x.imag, 4)))(M.full())
    return Qobj(arr)
```

- `pauli_map`: expected Heisenberg-picture mapping $(X\to, Y\to, Z\to)$

## Validation 1: Pulse Sequence vs Axis-Angle Target

For each Clifford, verify that the pulse sequence produces the same unitary as $R_{\hat{n}}(\theta) = e^{-i\theta(\hat{n}\cdot\vec{\sigma})/2}$ (up to global phase).

```python
print('Validation 1: Pulse sequence == axis/angle target (up to global phase)')
print('=' * 72)

all_pass_1 = True
for i, c in enumerate(cliffords):
    U_pulse = c['pulse']()
    if c['axis'] is not None:
        U_target = R_axis(c['axis'], c['angle'])
    else:
        U_target = si

    ok = phase_equiv(U_pulse, U_target)
    status = 'вњ“' if ok else 'вњ— FAIL'
    print(f"  C{i:2d}  {c['name']:6s}  {status}")
    if not ok:
        all_pass_1 = False
        print(f'       Pulse: {clean(U_pulse)}')
        print(f'       Target: {clean(U_target)}')

print('\n' + ('ALL PASSED вњ“' if all_pass_1 else 'SOME FAILED вњ—'))
```

## Validation 2: Heisenberg-Picture Pauli Mapping

For each Clifford $C$, verify that $C \sigma_\alpha C^\dagger$ matches the expected Pauli mapping.

```python
def identify_pauli(M, tol=1e-9):
    """Identify which В±Pauli a matrix is (or return None)."""
    for label, P in PAULIS.items():
        if (M - P).norm() < tol:
            return label
    return None

print('Validation 2: Pauli conjugation mapping')
print('=' * 72)

all_pass_2 = True
for i, c in enumerate(cliffords):
    U = c['pulse']()
    results = {}
    ok = True
    for pauli_name, pauli_op, expected_key in [('X', sx, 'x'), ('Y', sy, 'y'), ('Z', sz, 'z')]:
        conjugated = U * pauli_op * U.dag()
        label = identify_pauli(conjugated)
        results[pauli_name] = label
        if label != c[expected_key]:
            ok = False

    status = 'вњ“' if ok else 'вњ— FAIL'
    print(f"  C{i:2d}  {c['name']:6s}  Xв†’{results['X']:3s}  Yв†’{results['Y']:3s}  Zв†’{results['Z']:3s}  {status}")
    if not ok:
        all_pass_2 = False
        print(f"       Expected: Xв†’{c['x']}  Yв†’{c['y']}  Zв†’{c['z']}")

print('\n' + ('ALL PASSED вњ“' if all_pass_2 else 'SOME FAILED вњ—'))
```

## Validation 2b: Rx/Ry-Only Sequences Match Rz-Allowed Sequences

Verify that the `pulse_xy` (Rx/Ry only) decomposition equals the `pulse` (with Rz) decomposition up to global phase. Also verify the Pauli mapping is correct.

```python
print('Validation 2b: Rx/Ry-only sequences')
print('=' * 72)

all_pass_2b = True
for i, c in enumerate(cliffords):
    U_rz = c['pulse']()
    U_xy = c['pulse_xy']()

    equiv = phase_equiv(U_rz, U_xy)

    results = {}
    pauli_ok = True
    for pauli_name, pauli_op, expected_key in [('X', sx, 'x'), ('Y', sy, 'y'), ('Z', sz, 'z')]:
        conjugated = U_xy * pauli_op * U_xy.dag()
        label = identify_pauli(conjugated)
        results[pauli_name] = label
        if label != c[expected_key]:
            pauli_ok = False

    ok = equiv and pauli_ok
    status = 'вњ“' if ok else 'вњ— FAIL'
    print(f"  C{i:2d}  {c['name']:6s}  Xв†’{results['X']:3s}  Yв†’{results['Y']:3s}  Zв†’{results['Z']:3s}  phase_eq={'Y' if equiv else 'N'}  {status}")
    if not ok:
        all_pass_2b = False

print('\n' + ('ALL PASSED вњ“' if all_pass_2b else 'SOME FAILED вњ—'))
```

## Validation 3: Uniqueness вЂ” All 24 Are Distinct

Verify no two Cliffords are the same (up to global phase).

```python
print('Validation 3: Uniqueness check')
print('=' * 72)

unitaries = [c['pulse']() for c in cliffords]
duplicates = []
for i in range(24):
    for j in range(i + 1, 24):
        if phase_equiv(unitaries[i], unitaries[j]):
            duplicates.append((i, j))

if duplicates:
    print('DUPLICATES FOUND вњ—:')
    for i, j in duplicates:
        print(f'  C{i} ({cliffords[i]["name"]}) == C{j} ({cliffords[j]["name"]})')
else:
    print(f'All 24 Cliffords are distinct вњ“')
```

## Validation 4: Completeness вЂ” Generators Produce All 24

The single-qubit Clifford group is generated by $H$ and $S$. Verify that composing $H$ and $S$ can produce all 24 elements.

```python
print('Validation 4: Completeness via generators H, S')
print('=' * 72)

H_gate = cliffords[10]['pulse']()
S_gate = cliffords[8]['pulse']()

generated = {0: si}  # start with identity
queue = [si]
max_depth = 10

# BFS over products of H and S
for _ in range(max_depth):
    new_queue = []
    for U in queue:
        for gen in [H_gate, S_gate]:
            V = gen * U
            is_new = True
            for existing in generated.values():
                if phase_equiv(V, existing):
                    is_new = False
                    break
            if is_new:
                idx = len(generated)
                generated[idx] = V
                new_queue.append(V)
    queue = new_queue
    if len(generated) == 24:
        break

print(f'Generated {len(generated)} unique Cliffords from H and S')

# check every generated element matches one of our 24
matched = set()
for V in generated.values():
    for i, c in enumerate(cliffords):
        if phase_equiv(V, c['pulse']()):
            matched.add(i)
            break

if matched == set(range(24)):
    print('All 24 Cliffords are reachable from {H, S} вњ“')
else:
    missing = set(range(24)) - matched
    print(f'Missing Cliffords: {[cliffords[i]["name"] for i in missing]} вњ—')
```

## Validation 5: Group Closure

Verify that the product of any two Cliffords is again one of the 24.

```python
print('Validation 5: Group closure (24 Г— 24 = 576 products)')
print('=' * 72)

unitaries = [c['pulse']() for c in cliffords]
closure_ok = True
failures = 0

for i in range(24):
    for j in range(24):
        product = unitaries[i] * unitaries[j]
        found = False
        for k in range(24):
            if phase_equiv(product, unitaries[k]):
                found = True
                break
        if not found:
            closure_ok = False
            failures += 1
            if failures <= 5:
                print(f'  C{i} * C{j} not in group! вњ—')

if closure_ok:
    print('All 576 products land in the group вњ“')
else:
    print(f'{failures} products failed closure вњ—')
```

## Validation 6: Inverses Exist

For each Clifford, verify it has an inverse in the group.

```python
print('Validation 6: Inverses')
print('=' * 72)

unitaries = [c['pulse']() for c in cliffords]
all_pass_6 = True

for i in range(24):
    U_inv = unitaries[i].dag()
    found_inv = None
    for j in range(24):
        if phase_equiv(U_inv, unitaries[j]):
            found_inv = j
            break
    if found_inv is not None:
        print(f"  C{i:2d} ({cliffords[i]['name']:6s})вЃ»В№ = C{found_inv:2d} ({cliffords[found_inv]['name']})")
    else:
        print(f"  C{i:2d} ({cliffords[i]['name']:6s}) has no inverse in the set! вњ—")
        all_pass_6 = False

print('\n' + ('ALL PASSED вњ“' if all_pass_6 else 'SOME FAILED вњ—'))
```
